Programa que busca la palabra en google y la traduce y te da la pronunciacion, conetado a internet

```python
import tkinter as tk
from tkinter import messagebox
from googletrans import Translator
from gtts import gTTS
import os
```

```python

translator = Translator()

def traducir():
    palabra_ingles = entry.get()
    if palabra_ingles:
        traduccion = translator.translate(palabra_ingles, src='en', dest='es').text
        label_traduccion.config(text=traduccion)
    else:
        messagebox.showwarning("Entrada vacía", "Por favor, ingresa una palabra en inglés")

def pronunciar():
    palabra_ingles = entry.get()
    if palabra_ingles:
        tts = gTTS(text=palabra_ingles, lang='en')
        tts.save("temp.mp3")
        os.system("start temp.mp3" if os.name == "nt" else "mpg321 temp.mp3")
    else:
        messagebox.showwarning("Entrada vacía", "Por favor, ingresa una palabra en inglés")

# Configuración de la interfaz
tk_root = tk.Tk()
tk_root.title("Aprende Vocabulario en Inglés")

tk.Label(tk_root, text="Palabra en Inglés:").pack()
entry = tk.Entry(tk_root)
entry.pack()

btn_traducir = tk.Button(tk_root, text="Ver Significado", command=traducir)
btn_traducir.pack()

label_traduccion = tk.Label(tk_root, text="", font=("Arial", 14))
label_traduccion.pack()

btn_pronunciar = tk.Button(tk_root, text="Escuchar Pronunciación", command=pronunciar)
btn_pronunciar.pack()

tk_root.mainloop()
```

Programa que corre en local, que lee de un csv (palabra - traduccion) y que lee el audio de un archivo palabra.mp3

```python
import tkinter as tk
from tkinter import messagebox
import csv
import os

# Función para cargar las palabras y sus traducciones desde un archivo CSV
def cargar_diccionario(csv_file):
    diccionario = {}
    with open(csv_file, newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        for row in reader:
            if len(row) == 2:
                diccionario[row[0].strip().lower()] = row[1].strip()  # Guardar en un diccionario con claves en minúsculas
    return diccionario

# Cargar el diccionario desde el archivo CSV
diccionario = cargar_diccionario("vocabulario.csv")

# Función para traducir una palabra ingresada por el usuario
def traducir():
    palabra_ingles = entry.get().strip().lower()  # Obtener la palabra ingresada y convertirla a minúsculas
    if palabra_ingles in diccionario:
        label_traduccion.config(text=diccionario[palabra_ingles])  # Mostrar la traducción en la interfaz
    else:
        messagebox.showwarning("Palabra no encontrada", "No se encontró la traducción en la base de datos.")

# Función para reproducir el audio de la pronunciación de la palabra
def pronunciar():
    palabra_ingles = entry.get().strip().lower()
    audio_file = f"audio/{palabra_ingles}.mp3"  # Ruta del archivo de audio
    if os.path.exists(audio_file):
        os.system(f"start {audio_file}" if os.name == "nt" else f"mpg321 {audio_file}")  # Reproducir el audio en Windows o Linux
    else:
        messagebox.showwarning("Audio no encontrado", "No se encontró el archivo de audio para esta palabra.")

# Configuración de la ventana principal
tk_root = tk.Tk()
tk_root.title("Aprende Vocabulario en Inglés")

# Etiqueta y campo de entrada para la palabra en inglés
tk.Label(tk_root, text="Palabra en Inglés:").pack()
entry = tk.Entry(tk_root)
entry.pack()

# Botón para traducir la palabra
btn_traducir = tk.Button(tk_root, text="Ver Significado", command=traducir)
btn_traducir.pack()

# Etiqueta donde se mostrará la traducción
label_traduccion = tk.Label(tk_root, text="", font=("Arial", 14))
label_traduccion.pack()

# Botón para reproducir la pronunciación de la palabra
btn_pronunciar = tk.Button(tk_root, text="Escuchar Pronunciación", command=pronunciar)
btn_pronunciar.pack()

# Ejecutar la aplicación
tk_root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
import csv
import os

# Función para cargar las palabras y sus traducciones desde un archivo CSV
def cargar_diccionario(csv_file):
    lista_palabras = []
    with open(csv_file, newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        for row in reader:
            if len(row) == 2:
                lista_palabras.append((row[0].strip().lower(), row[1].strip()))  # Guardar como lista de tuplas (inglés, español)
    return lista_palabras

# Cargar la lista de palabras desde el archivo CSV
diccionario = cargar_diccionario("vocabulario.csv")
indice_actual = 0  # Índice para recorrer la lista de palabras

def mostrar_palabra():
    global indice_actual
    if diccionario:
        palabra_ingles, traduccion = diccionario[indice_actual]
        label_palabra.config(text=palabra_ingles)
        label_traduccion.config(text="???")  # Ocultar la traducción

def mostrar_traduccion():
    global indice_actual
    if diccionario:
        _, traduccion = diccionario[indice_actual]
        label_traduccion.config(text=traduccion)

def pronunciar():
    global indice_actual
    if diccionario:
        palabra_ingles, _ = diccionario[indice_actual]
        audio_file = f"audio/{palabra_ingles}.mp3"  # Ruta del archivo de audio
        if os.path.exists(audio_file):
            os.system(f"start {audio_file}" if os.name == "nt" else f"mpg321 {audio_file}")  # Reproducir el audio en Windows o Linux
        else:
            messagebox.showwarning("Audio no encontrado", "No se encontró el archivo de audio para esta palabra.")

def siguiente_palabra():
    global indice_actual
    if diccionario:
        indice_actual = (indice_actual + 1) % len(diccionario)  # Avanzar a la siguiente palabra en bucle
        mostrar_palabra()

# Configuración de la ventana principal
tk_root = tk.Tk()
tk_root.title("Aprende Vocabulario en Inglés")

# Etiqueta para mostrar la palabra en inglés
label_palabra = tk.Label(tk_root, text="", font=("Arial", 16, "bold"))
label_palabra.pack()

# Botón para mostrar la traducción
btn_traducir = tk.Button(tk_root, text="Ver Significado", command=mostrar_traduccion)
btn_traducir.pack()

# Etiqueta donde se mostrará la traducción
label_traduccion = tk.Label(tk_root, text="", font=("Arial", 14))
label_traduccion.pack()

# Botón para reproducir la pronunciación
btn_pronunciar = tk.Button(tk_root, text="Escuchar Pronunciación", command=pronunciar)
btn_pronunciar.pack()

# Botón para pasar a la siguiente palabra
btn_siguiente = tk.Button(tk_root, text="Siguiente Palabra", command=siguiente_palabra)
btn_siguiente.pack()

# Mostrar la primera palabra
mostrar_palabra()

# Ejecutar la aplicación
tk_root.mainloop()
```

```python
#programa para sacar los mp3 de la pronunciacion
import os
import pandas as pd
from gtts import gTTS

# Nombre del archivo CSV
csv_file = "vocabulario.csv"

# Crear la carpeta de audio si no existe
audio_folder = "audio"
os.makedirs(audio_folder, exist_ok=True)

# Cargar el archivo CSV
df = pd.read_csv(csv_file)

# Iterar sobre cada palabra y generar el audio
for palabra in df.iloc[:, 0]:  # Tomar solo la primera columna
    palabra = str(palabra).strip()  # Convertir a string y limpiar espacios
    if palabra:
        audio_path = os.path.join(audio_folder, f"{palabra}.mp3")
        tts = gTTS(text=palabra, lang="en")  # Generar audio en inglés
        tts.save(audio_path)
        print(f"Guardado: {audio_path}")

print("¡Proceso completado!")
```
