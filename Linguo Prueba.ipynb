Programa que busca la palabra en google y la traduce y te da la pronunciacion, conetado a internet

```python
import tkinter as tk
from tkinter import messagebox
from googletrans import Translator
from gtts import gTTS
import os
```

```python

translator = Translator()

def traducir():
    palabra_ingles = entry.get()
    if palabra_ingles:
        traduccion = translator.translate(palabra_ingles, src='en', dest='es').text
        label_traduccion.config(text=traduccion)
    else:
        messagebox.showwarning("Entrada vacía", "Por favor, ingresa una palabra en inglés")

def pronunciar():
    palabra_ingles = entry.get()
    if palabra_ingles:
        tts = gTTS(text=palabra_ingles, lang='en')
        tts.save("temp.mp3")
        os.system("start temp.mp3" if os.name == "nt" else "mpg321 temp.mp3")
    else:
        messagebox.showwarning("Entrada vacía", "Por favor, ingresa una palabra en inglés")

# Configuración de la interfaz
tk_root = tk.Tk()
tk_root.title("Aprende Vocabulario en Inglés")

tk.Label(tk_root, text="Palabra en Inglés:").pack()
entry = tk.Entry(tk_root)
entry.pack()

btn_traducir = tk.Button(tk_root, text="Ver Significado", command=traducir)
btn_traducir.pack()

label_traduccion = tk.Label(tk_root, text="", font=("Arial", 14))
label_traduccion.pack()

btn_pronunciar = tk.Button(tk_root, text="Escuchar Pronunciación", command=pronunciar)
btn_pronunciar.pack()

tk_root.mainloop()
```
